```python
#load python packages
import os
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
path = os.getcwd()
file_name = 'fake_job_postings_no_outliers.csv'
path += "/../data/interim/"+file_name
df = pd.read_csv(path, index_col='job_id')
```

Categorical features have already been dummified in the data wrangling step. Outliers of numerical columns were removed in EDA step. I will now split into test/train sets and standardize numerical columns

```python
df.head()
```

```python
from sklearn.model_selection import train_test_split

X = df.drop('fraudulent',axis=1)
y = df['fraudulent']
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=1, stratify=y)
```

```python
from sklearn.preprocessing import StandardScaler

numeric_cols = ['company_profile_length', 'description_length', 'requirements_length']
scaler = StandardScaler()
X_train_scaled = X_train.copy() #setting up the categorical dummy columns which will not be scaled
X_test_scaled = X_test.copy()
X_train_scaled.loc[:,numeric_cols] = scaler.fit_transform(X_train.loc[:,numeric_cols]) #scaling numeric columns
X_test_scaled.loc[:,numeric_cols] = scaler.transform(X_test.loc[:,numeric_cols])
```

```python
X_test_scaled[numeric_cols]
```

```python
X_train_scaled.to_csv('../data/processed/fake_job_postings_X_train.csv')
X_test_scaled.to_csv('../data/processed/fake_job_postings_X_test.csv')
y_train.to_csv('../data/processed/fake_job_postings_y_train.csv')
y_test.to_csv('../data/processed/fake_job_postings_y_test.csv')
```
